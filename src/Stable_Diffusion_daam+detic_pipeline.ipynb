The following is an implementation of the single image pipeline. You can try it out to get an idea of the outputs obtained and check the comparison with the `Detic` baseline.

```python
%%bash
source activate env
python

import os 
import json
import cv2
import numpy as np
import torch
from PIL import Image
from daam import trace, set_seed
from diffusers import StableDiffusionPipeline

import nltk
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

prompt = "little girl wearing pink dress standing in wooden cabin"
stop_words = set(stopwords.words('english'))

# Extracting nouns from prompt
tokenized = sent_tokenize(prompt)
nouns = []
for i in tokenized:
    wordsList = word_tokenize(i)
 
    wordsList = [w for w in wordsList if not w in stop_words]
    tagged = pos_tag(wordsList)
 
    nouns = [word for word, tag in tagged if tag == 'NN']

image_id = "1"
gen = set_seed(1)

bbox_path = "bboxes"
bbox_output_path = "bbox_output"
json_path = "json_path_matches"
thresh_path = "thresh_image"
genimg_path = "generated_image"
heatmaps_path = "heatmaps"
daam_output_path = "daam_output"
contour_points_path = "contour_points"

json_dict = dict()

model_id = 'stabilityai/stable-diffusion-2'
device = 'cuda'

pipe = StableDiffusionPipeline.from_pretrained(model_id, use_auth_token=True, torch_dtype=torch.float16, revision="fp16")
pipe = pipe.to(device)

with torch.cuda.amp.autocast(dtype=torch.float16), torch.no_grad():
    with trace(pipe) as tc:
        out = pipe(prompt, num_inference_steps=10, generator=gen)       # generate output from Stable Diffusion pipe
        img = out.images[0]   

        image_path = genimg_path + os.sep + image_id + os.sep + "generated_image.png"
        if not os.path.exists(genimg_path + os.sep + image_id + os.sep) : 
          os.mkdir(genimg_path + os.sep + image_id + os.sep)
        img.save(image_path)                                            # save generated image

        glob_heat_map = tc.compute_global_heat_map(prompt)              # generate global heat map

        for word in nouns : 
          heat_map = glob_heat_map.compute_word_heat_map(word)          # compute word-wise heatmap

          out_file = daam_output_path + os.sep + image_id + os.sep + word + os.sep + "daam_output.png"
          if not os.path.exists(daam_output_path + os.sep + image_id + os.sep + word + os.sep) : 
            os.makedirs(daam_output_path + os.sep + image_id + os.sep + word + os.sep)

          # create a new black image of same dimensions as original image, and overlay heat map on it
          img_shape = np.array(img).shape
          new_img = Image.fromarray(np.zeros(img_shape, dtype = np.uint8))
          heat_map.plot_overlay(new_img, out_file=out_file)         # overlay word-wise heatmap on generated img and save it

          real_image = cv2.imread(image_path)                   # read original (generated) image 
          real_image = cv2.resize(real_image, (369, 368),       # shrink it to the size of daam output image
               interpolation = cv2.INTER_AREA)
          print("Image dimensions for original image : {}, {}".format(real_image.shape[0], real_image.shape[1]))
          image =  cv2.imread(out_file)                         # read word-wise heat map image
          image = image[59:427, 144:513]                        # crop out the relevant portion from image
          print("Image dimensions for word heatmap : {}, {}".format(image.shape[0], image.shape[1]))
      
          # obtaining bboxes from heatmaps
          gray =  cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
          thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

          # saving thresholded images
          thr_path = thresh_path + os.sep + image_id + os.sep + word + os.sep + "thresh.png"
          os.makedirs(thresh_path + os.sep + image_id + os.sep + word + os.sep, exist_ok=True)
          cv2.imwrite(thr_path, thresh)

          # finding contours from thresholded images
          cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
          cnts = cnts[0] if len(cnts) == 2 else cnts[1]
          print("Length of cnts for {} : {}".format(word, len(cnts)))

          # draw bbox rectangles on image
          bbox_image = image.copy()
          bb_out_path = bbox_output_path + os.sep + image_id + os.sep + word + os.sep + "daam_bbox.png"
          bb_path = bbox_path + os.sep + image_id + os.sep + word + os.sep
          os.makedirs(bbox_output_path + os.sep + image_id + os.sep + word + os.sep, exist_ok=True)   # create path for saving bbox_output
          os.makedirs(bbox_path + os.sep + image_id + os.sep + word + os.sep, exist_ok=True)        # create path for saving bboxes

          num_bboxes = 0
          word_bbox_paths = []
          for c in cnts:
              x, y, w, h = cv2.boundingRect(c)
                                                            # threshold for bbox size set bcoz minute disturbances in heat-map
              if w > 50 and h > 50:                       # are also detected as contours, and will be drawn on image
                cv2.rectangle(real_image, (x, y), (x + w, y + h), (0,255,0), 2)
                cv2.putText(real_image, word, (x+10, y+20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 2)
                print("x : {}, y : {}, w : {}, h : {}".format(x, y, w, h))
                bbox_dims = np.array([x, y, w, h])

                # save bbox coordinates for each word
                np.save(bb_path + str(num_bboxes), bbox_dims)
                word_bbox_paths.append(bb_path + str(num_bboxes)+".npy")
                num_bboxes += 1

          json_dict[word] = word_bbox_paths
          cv2.imwrite(bb_out_path, real_image)

          t_path = thresh_path +  os.sep + image_id + os.sep + word + os.sep + "thresh.png"

          if not os.path.exists(thresh_path +  os.sep + image_id + os.sep + word + os.sep):
            os.makedirs(thresh_path + os.sep + image_id + os.sep + word + os.sep)

          cv2.imwrite(t_path, thresh)

# save bbox paths dictionary in JSON format
with open("bbox_paths.json", "w") as f:
    json.dump(json_dict, f)
```

###CREATING ANNOTATED DATA

The captions used for generating images using Stable Diffusion are obtained from the `flickr30k_all.json` file, which contains a total of 150k captions for 30k images in the Flickr30k dataset.

- Model used : `stabilityai/stable-diffusion-2`
- Only stop-word removal has been used for obtaining words from captions.
- Due to computational constraints, the maximum number of generated images is limited by the `max_iter` parameter, which is set to 30. This can be set to any value <= 150k, the total number of captions in the dataset.


###DIRECTORY STRUCTURE

> `data/bboxes/{word}/{image_id}.npy` : contains bbox annotations for ALL instances of word in image with `image_id`.

> `data/heatmaps/{word}/{image_id}.pt` : contains attention maps (heatmaps) for word in image with `image_id`.

> `data/bboxes/{word}/{image_id}.png` : contains heatmaps overlayed on dark background for word in image with `image_id`.

> `data/images/{image_id}.png` : contains generated images.








```python
%%bash
source activate env
cd data
python 

import os 
import json
import cv2
import numpy as np
import torch
from PIL import Image
from daam import trace, set_seed
from diffusers import StableDiffusionPipeline

import nltk
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

# function to preprocess tokens of a prompt
def preprocess_tokens(tokens): 
  wordsList = [w for w in tokens if not w in stop_words]
  tokens_ = list(set(wordsList))            # remove repeated words

#   tagged = pos_tag(wordsList)               # apply POS tagging
#   tokens_ = [word for word, tag in tagged if tag == 'NN']     # extract nouns
  return tokens_

# Load dataset captions
print("Loading Flickr30k captions ...")
captions_path = '/content/drive/MyDrive/AIISC/data/flickr30k_all.json'
captions_dict = dict()

with open(captions_path, "r") as f:
  captions_dict = json.load(f)

# define directory names
bbox_path = "bboxes"
genimg_path = "images"
heatmaps_path = "heatmaps"
heatmap_imgs_path = "heatmap_imgs"

gen = set_seed(0)
image2word = dict()           # dictionary to store generated image id with corresponding list of caption words

model_id = 'stabilityai/stable-diffusion-2'
device = 'cuda'

# Load Stable Diffusion model pipeline
print("Loading model ...")
pipe = StableDiffusionPipeline.from_pretrained(model_id, use_auth_token=True, torch_dtype=torch.float16, revision="fp16")
pipe = pipe.to(device)
print("Model used : ", model_id)

# Temporary flag variable to limit no. of iterations
is_looping = True
max_iter = 30

for i, img_dict in enumerate(captions_dict['images']):
  
  if not is_looping:        # flag variable check
    break
  for j, sentence_dict in enumerate(img_dict['sentences']):

    prompts_per_img = len(img_dict['sentences'])
    idx = i * prompts_per_img + j           # compute image_id
    image_id = str(idx)

    # flag variable check
    if idx >= max_iter:
        is_looping = False
        break

    tokens = sentence_dict['tokens']
    prompt = sentence_dict['raw']
    tokens = [token.lower() for token in tokens]
    tokens = preprocess_tokens(tokens)
    image2word[image_id] = tokens           # save image_id with corresp caption tokens

    print("Generating image for image id {} ...".format(image_id))

    with torch.cuda.amp.autocast(dtype=torch.float16), torch.no_grad():
      with trace(pipe) as tc:
        out = pipe(prompt, num_inference_steps=10, generator=gen)       # generate output from Stable Diffusion pipe
        img = out.images[0]   

        # save generated image
        image_path = genimg_path + os.sep + image_id + ".png"
        if not os.path.exists(genimg_path) : 
          os.makedirs(genimg_path)
        img.save(image_path)                                            

        glob_heat_map = tc.compute_global_heat_map(prompt)              # generate global heat map

        for word in tokens : 
          try:
            heat_map = glob_heat_map.compute_word_heat_map(word)          # compute word-wise heatmap
          except:
            continue

          # save heatmap (tensor) as .pt file
          hmap_path = heatmaps_path + os.sep + word + os.sep + image_id + ".pt"
          if not os.path.exists(heatmaps_path + os.sep + word) : 
            os.makedirs(heatmaps_path + os.sep + word)
          torch.save(heat_map, hmap_path)
          
          hmap_img_path = heatmap_imgs_path + os.sep + word + os.sep + image_id + ".png"
          if not os.path.exists(heatmap_imgs_path + os.sep + word) : 
            os.makedirs(heatmap_imgs_path + os.sep + word)

          # create a new black image of same dimensions as original image, and overlay heat map on it
          img_shape = np.array(img).shape
          mono_img = Image.fromarray(np.zeros(img_shape, dtype = np.uint8))
          heat_map.plot_overlay(mono_img, out_file=hmap_img_path)                       # overlay word-wise heatmap on generated img and save it

          real_image = cv2.imread(image_path)                   # read original (generated) image 
          real_image = cv2.resize(real_image, (369, 368),       # shrink it to the size of daam output image
               interpolation = cv2.INTER_AREA)

          image = cv2.imread(hmap_img_path)                     # read word-wise heat map image
          # print("Image shape : ", image.shape)
          image = image[59:427, 144:513]                        # crop out the relevant portion from image
      
          # obtaining bboxes from heatmaps
          gray =  cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
          thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

          # finding contours from thresholded images
          cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
          cnts = cnts[0] if len(cnts) == 2 else cnts[1]
          # print("Length of cnts : ", len(cnts))

          # define path for bbox
          bb_path = bbox_path + os.sep + word + os.sep + image_id + ".npy"
          os.makedirs(bbox_path + os.sep + word + os.sep, exist_ok=True)        

          # draw bbox rectangles on image
          bbox_image = image.copy()

          word_bboxes = []
          print("Checking contours ...")

          for c in cnts:
              x, y, w, h = cv2.boundingRect(c)
              # print("x : {}, y : {}, w : {}, h : {}".format(x, y, w, h))

              bbox_dims = np.array([x, y, w, h])
              word_bboxes.append(bbox_dims)

          # save bbox coordinates for each word
          word_bboxes = np.array(word_bboxes)
          np.save(bb_path, word_bboxes)

# save image2word dictionary in JSON format
with open("image2word.json", "w") as f:
    json.dump(image2word, f)    
```

### INFERENCE

For a given word, the inference pipeline performs `text-based image retrieval` and returns all images which contain instances of the word, along with the bboxes for those instances.

> `data/output/{word}/{image_id}.png` : Image with `image_id` annotated with bboxes for all instances of the search word.

```python
%%bash
source activate env
cd data
python

import os
import cv2
import numpy as np

word = "girl"

cwd = os.getcwd()
bbox_path = cwd + os.sep + "bboxes" + os.sep + word
img_path = cwd + os.sep + "images" + os.sep
bboxes_list = os.listdir(bbox_path)

output_path = cwd + os.sep + "output" + os.sep + word
os.makedirs(output_path, exist_ok=True)

if len(bboxes_list) > 0 :
    for bbox_filename in bboxes_list:
        image_id = bbox_filename[:-4]
        bbox_file_path = bbox_path + os.sep + bbox_filename
        print("bbox file path : ", bbox_file_path)
        bboxes = np.load(bbox_file_path)
        print("bboxes : ",bboxes)

        image_path = img_path + image_id + ".png"
        print("Matching image : ", image_path)
        image = cv2.imread(image_path)
        image = cv2.resize(image, (369, 368),
              interpolation = cv2.INTER_AREA)

        for bbox in bboxes:
            x, y, w, h = list(bbox)
            print("x : {}, y : {}, w : {}, h : {}".format(x, y, w, h))
            cv2.rectangle(image, (x, y), (x + w, y + h), (0,255,0), 2)
            cv2.putText(image, word, (x+10, y+20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 2)

        out_path = output_path + os.sep + image_id + ".png"
        cv2.imwrite(out_path, image)
```
